```python
res_red = pd.read_csv("res_mult_20000.csv", delimiter=";")

res_full = pd.read_csv("res_mult_40000.csv", delimiter=";")
```

```python
plt.plot(res_red["Amount_of_threads"], res_red["Speedup"], markersize=7, marker='o')
plt.plot(res_full["Amount_of_threads"], res_full["Speedup"], markersize=7, marker='o')
plt.plot([0,40], [0,40])
plt.grid()
plt.xlabel('Threads')
plt.ylabel('Speedup')
plt.legend(["len_20000", "len_40000"])
plt.show()
```

```python
import matplotlib.pyplot as plt

# Создаем данные для таблицы
data = [
    ["M=N", "Количество потоков"],
    ["", 2, 4, 6, 8],
    ["", "T1", "T2", "S2", "T4", "S4", "T6", "S6", "T8", "S8"],
    ["20000"],
    ["40000"]
    
]

# Создаем фигуру и оси
fig, ax = plt.subplots(figsize=(8, 4))

# Создаем таблицу
table = ax.table(cellText=data,
                 loc='center',
                 cellLoc='center',
                 colWidths=[0.2, 1])

# Настраиваем таблицу
table.auto_set_font_size(False)
table.set_fontsize(12)
table.scale(1, 1.5)  # Увеличиваем высоту ячеек

# Убираем оси
ax.axis('off')

# Добавляем заголовок
plt.title("Пример таблицы с использованием Matplotlib", pad=20)

# Отображаем таблицу
plt.show()
```
